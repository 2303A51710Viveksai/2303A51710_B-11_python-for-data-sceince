```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
df = pd.DataFrame({
    "price": np.random.randint(50000, 500000, 100),
    "area": np.random.randint(600, 4000, 100),
    "rooms": np.random.randint(1, 10, 100)
})

sns.set(style="whitegrid", palette="mako")
sns.pairplot(df, diag_kind="kde")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

titanic = sns.load_dataset("titanic")
titanic = titanic[["age", "fare", "pclass", "survived"]].dropna()
sns.set(style="whitegrid", context="talk", palette="deep")
pair = sns.pairplot(titanic, vars=["age", "fare", "pclass"], hue="survived", diag_kind="kde", height=3, plot_kws={"alpha":0.7,"s":80,"edgecolor":"k"})
pair.fig.suptitle("Titanic Passenger Data: Relationships Among Age, Fare, and Class", fontsize=18, y=1.03)
plt.show()
```
